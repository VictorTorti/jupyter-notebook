```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy as sp

import pandas as pd

indice = ['uno','dos','tres','cuatro','cinco']
columnas = ['Raza','Puntos']
lista = [['Caniche',8.1],['Bulldog',7.3],['ChowChow',7.6],['Chihuahua',9.0],['Labrador',9.3]]


print('----------------------------')
print('--- Apartado 1            --')
print('----------------------------')

df = pd.DataFrame(lista, columns=columnas, index=indice)
print(df)

print('----------------------------')
print('--- Apartado 2            --')
print('----------------------------')

puntos_ana = [61,75,82,95,99]
df['PuntosAna'] = puntos_ana

print(df)

print('----------------------------')
print('--- Apartado 3            --')
print('----------------------------')

df['PuntosAna'] = df['PuntosAna'] / 10

print(df)

print('----------------------------')
print('--- Apartado 4            --')
print('----------------------------')

lista2 = [['Samoyedo', 9.2, 8.9], ['Pinscher', 8.1, 6.7]]

df2 = pd.DataFrame(lista2, columns=['Raza', 'Puntos', 'PuntosAna'], index=['seis', 'siete'])

df = pd.concat([df, df2])

print(df)

print('----------------------------')
print('--- Apartado 5            --')
print('----------------------------')

df['Media'] = (df['Puntos'] + df['PuntosAna']) / 2

print(df)

print('----------------------------')
print('--- Apartado 6            --')
print('----------------------------')

df = df.drop('dos')
print(df)

print('----------------------------')
print('--- Apartado 7            --')
print('----------------------------')

df = df.iloc[2:5]

print(df)

print('----------------------------')
print('--- Apartado 8            --')
print('----------------------------')

print(df.describe())
```
